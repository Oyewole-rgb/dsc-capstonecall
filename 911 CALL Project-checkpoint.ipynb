```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
calls=pd.read_csv('911.csv')
calls.head()
```

```python
calls.info()
```

```python
calls.head()
```

```python
calls['zip'].value_counts().head(5)
```

```python
calls['twp'].value_counts().head(5)
```

```python
calls['title'].nunique()
```

```python
calls['Reason']=calls['title'].apply(lambda title: title.split(':')[0])
```

```python
calls['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=calls)
```

```python
type(calls['timeStamp'].iloc[0])
```

```python
time=calls['timeStamp'].iloc[0]
time.hour
```

```python
time.dayofweek
```

```python
calls['Hour']=calls['timeStamp'].apply(lambda time:time.hour)
calls['Month']=calls['timeStamp'].apply(lambda time:time.month)
calls['Day of Week']=calls['timeStamp'].apply(lambda time:time.dayofweek)

dmap={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
calls['Day of Week']=calls['Day of Week'].map(dmap)
```

```python
calls.head()
```

```python
sns.countplot(x='Day of Week',data=calls,hue='Reason')
plt.legend(bbox_to_anchor=(1.05,1), loc=2,borderaxespad=0.)
```

```python
sns.countplot(x='Month',data=calls,hue='Reason')
plt.legend(bbox_to_anchor=(1.05,1), loc=2,borderaxespad=0.)
```

```python
byMonth = calls.groupby('Month').count()
byMonth.head()
```

```python
byMonth['lat'].plot()
```

```python
sns.countplot(x='Month',data=calls)
plt.legend(bbox_to_anchor=(1.05,1), loc=2,borderaxespad=0.)
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
time=calls['timeStamp'].iloc[0]
calls['Date']=calls['timeStamp'].apply(lambda time:time.date())
```

```python
calls.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
calls[calls['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
calls[calls['Reason']=='Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
calls[calls['Reason']=='EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour=calls.groupby(by=['Day of Week', 'Hour']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

```python
dayMonth=calls.groupby(by=['Day of Week', 'Month']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='coolwarm')
```
